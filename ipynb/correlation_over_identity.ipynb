```python
from Bio.SeqIO import parse as parse_seq_file

recs = parse_seq_file('../seq/16S.ungap.afn', 'fasta')

#i = iter(recs)
#a = i.next()
#b = i.next()
```

```python
recs_list = list(recs)
```

```python
def perc_identity(seq1, seq2):
    """Calculate the percent nucleotide identity between two sequences.
    
    Does not count shared gaps.
    Counts unshared gaps.
    """
    seq1 = asarray(seq1)
    seq2 = asarray(seq2)
    in_one = (seq1 != '-') | (seq2 != '-')
    identical = (seq1 == seq2)
    return float(sum(identical & in_one)) / sum(in_one)

#%timeit(perc_identity(a, b))
```

```python
def chunks(iterator, n, part_last=False):
    out = []
    for i, item in enumerate(iterator):
        out.append(item)
        if (i != 0) and (i % n):
            yield out
            out = []
    if out and part_last:
        yield out
```

```python
copies_map = {}
with open('../meta/16S.tsv') as handle:
    for line in handle:
        fields = line.split()
        key = fields[0]
        try:
            value = int(fields[3])
        except ValueError:
            continue
        else:
            if key in copies_map:
                assert copies_map[key] == value
            else:
                copies_map[key] = value
                
genome_map = {}
with open('../meta/16S.tsv') as handle:
    for line in handle:
        fields = line.split()
        key = fields[0]
        try:
            value = int(fields[1])
        except ValueError:
            continue
        else:
            if key in genome_map:
                assert genome_map[key] == value
            else:
                genome_map[key] = value
```

```python
# Takes about 10 minutes to run.

import random
from pandas import DataFrame

n = 10000
table = []
counter = 0
for seq1, seq2 in chunks(random.sample(recs_list, n), 2):
    table.append((seq1.name, seq2.name,
                  genome_map[seq1.name], genome_map[seq2.name],
                  copies_map[seq1.name], copies_map[seq2.name],
                  perc_identity(seq1, seq2)))


transpose = zip(*table)
data = DataFrame(dict(leaf1 = transpose[0],
                      leaf2 = transpose[1],
                      genome1 = transpose[2],
                      genome2 = transpose[3],
                      copy1 = transpose[4],
                      copy2 = transpose[5],
                      perc_ident = transpose[6]))
```

```python
import random

def random_combination(population, k, n=-1):
    i = 0
    while True:
        yield random.sample(population, k)
        i += 1
        if i == n:
            break
```

```python
import statsmodels.api as sm
from scipy.stats import t
from pandas import DataFrame as df

window_size = 0.01
dists = linspace(0.5, 1, 200)
columns = ['dist_center', 'dist_wind_low', 'dist_wind_hi',
           'R2', 'r', 'beta', 'beta_ci_low', 'beta_ci_hi',
           'ssr', 'n', 'df_resid', 'predict1', 'predict1_std']
table = DataFrame(index=range(len(dists)), columns=columns, dtype=float64)
table.dist_center = dists
table.dist_wind_low = table.dist_center - window_size / 2
table.dist_wind_hi = table.dist_center + window_size / 2
for i in range(len(dists)):
    sub_data = data2[(data2.perc_ident > table.dist_wind_low[i]) & \
                     (data2.perc_ident < table.dist_wind_hi[i])]
    if len(sub_data) < 2:
        continue
    lm = sm.OLS(sub_data['copy1'], sub_data['copy2'])
    fit = lm.fit()
    table.R2[i] = fit.rsquared
    table.r[i] = corrcoef(sub_data['copy1'], sub_data['copy2'])[0,1]
    table.beta[i] = fit.params[0]
    ci = fit.conf_int()
    table.beta_ci_low[i] = ci[0]
    table.beta_ci_hi[i] = ci[1]
    table.ssr[i] = fit.ssr
    table.n[i] = fit.nobs
    table.df_resid[i] = fit.df_resid
    table.predict1[i] = fit.predict(1)
    table.predict1_std[i] = sqrt(fit.mse_resid)
table['predict1_low'] = table.predict1 - (t.ppf(0.975, table.df_resid) * table.predict1_std)
table['predict1_hi'] = table.predict1 + (t.ppf(0.975, table.df_resid) * table.predict1_std)
    
fig, ax1 = plt.subplots()
ax1.plot(table.dist_center, table.r, label='$r$')
ax1.plot(table.dist_center, table.predict1, label='Prediction')
ax1.fill_between(table.dist_center, table.predict1_low, table.predict1_hi, alpha=0.2)
ax1.fill_between(dists, ci_low, ci_hi, lw=0, alpha=0.2, label="95% Confidence Band", color="green")
ax1.plot(dists, coefs, label="$\\hat{\\beta}$", lw=2)
ax1.plot(dists, rs, label='$r$', lw=2)
ax1.axhline(1, ls="--", color="k")
ax1.set_xlabel("Minimum Percent Identity")
ax1.legend(loc="upper right")
ax1.set_xlim((min(dists), max(dists)))
ax1.invert_xaxis()
ax2 = ax1.twinx()
ax2.hist(data2.perc_ident, bins=dists, normed=True, alpha=0.15, color='k', lw=0)
ax2.set_ylabel("Probability Density")
ax2.set_yticklabels([])
ax2.set_xlim((min(dists), max(dists)))
ax2.invert_xaxis()
```
